```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Bharathwaj\Sports\IPL_ball_by_ball_updated.csv')
```

```python
all_batsman = df['striker'].unique()

print(all_batsman)
```

```python
batsman_df = df[(df['bowler'] == 'JA Richardson')]
```

```python
batsman_df.tail(50)
```

```python
final_df = pd.DataFrame()
```

```python
for batsman in all_batsman:
    batsman_df = df[(df['striker'] == batsman)]
    batsman_df.loc[:, 'dots']= (batsman_df['runs_off_bat'] == 0).astype(int)
    batsman_df.loc[:, 'fours'] = (batsman_df['runs_off_bat'] == 4).astype(int)
    batsman_df.loc[:, 'sixes'] = (batsman_df['runs_off_bat'] == 6).astype(int)
    batsman_stats = batsman_df.groupby('bowler').agg({
    'ball': 'count',
    'runs_off_bat': 'sum',
    'player_dismissed': 'count',
    'dots': 'sum',
    'fours': 'sum',     
    'sixes': 'sum',
}).reset_index()
# Renaming columns for clarity
    batsman_stats['batsman'] = batsman
    batsman_stats.columns = ['bowler', 'balls_faced', 'runs_scored', 'times_dismissed', 'dots', '4s', '6s', 'batsman']
    final_df = final_df.append(batsman_stats, ignore_index=True)
```

```python
final_df.head(30)
```

```python
final_df_sorted = final_df.sort_values(by='4s', ascending=False)
final_df_sorted.head(30)
```

```python
final_df_sorted = final_df.sort_values(by='runs_scored', ascending=False)
final_df_sorted.head(30)
```

```python
final_df_sorted = final_df.sort_values(by='balls_faced', ascending=False)
final_df_sorted.head(30)
```

```python
final_df_sorted = final_df.sort_values(by='times_dismissed', ascending=False)
final_df_sorted.head(50)
```
