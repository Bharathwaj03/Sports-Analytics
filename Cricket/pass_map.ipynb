```python
import pandas as pd
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df = pd.read_csv(r"C:\\Users\\Lenovo\\Downloads\\match2.csv")
```

```python
df.head()
```

```python
dfPass = df[df['type']=='Pass']
```

```python
dfPass = dfPass[dfPass['outcomeType']=="Successful"]
```

```python
MC_pass = dfPass[dfPass['teamId']=="ManchesterCity"]
```

```python
pitch = mplsoccer.Pitch(pitch_type='opta', pitch_color ='grass', line_color='white',stripe=True)
fig,ax= pitch.draw(figsize=(18, 10))

pitch.arrows(MC_pass['x'],MC_pass['y'],MC_pass['endX'],MC_pass['endY'],color = '#4dd0e1',alpha= 0.7, width =2,ax=ax)
plt.show()
```

```python
MC_pass = dfPass[dfPass['teamId']=="ManchesterCity"]
pitch = mplsoccer.Pitch(pitch_type='opta', pitch_color ='green', line_color='white')
fig,ax= pitch.draw(figsize=(18, 10))

pitch.arrows(MC_pass['x'],MC_pass['y'],MC_pass['endX'],MC_pass['endY'],color = '#4dd0e1',comet='true' ,ax=ax)
```
