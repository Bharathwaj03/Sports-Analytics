```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Bharathwaj\Sports\IPL_ball_by_ball_updated.csv')
```

```python
batting_team = df[(df['batting_team'] == 'Mumbai Indians')]
```

```python
batting_team
```

```python
all_batsman = batting_team['striker'].unique()

print(all_batsman)
```

```python
final_df = pd.DataFrame()
```

```python
for batsman in all_batsman:
    batsman_df = batting_team[(batting_team['striker'] == batsman)]
    batsman_df.loc[:, 'dots']= (batsman_df['runs_off_bat'] == 0).astype(int)
    batsman_df.loc[:, 'fours'] = (batsman_df['runs_off_bat'] == 4).astype(int)
    batsman_df.loc[:, 'sixes'] = (batsman_df['runs_off_bat'] == 6).astype(int)
    batsman_stats = batsman_df.groupby('striker').agg({
    'ball': 'count',
    'runs_off_bat': 'sum',
    'player_dismissed': 'count',
    'dots': 'sum',
    'fours': 'sum',     
    'sixes': 'sum',
}).reset_index()
# Renaming columns for clarity
    batsman_stats['batsman'] = batsman
    batsman_stats.columns = ['bowler', 'balls_faced', 'runs_scored', 'times_dismissed', 'dots', '4s', '6s', 'batsman']
    final_df = final_df.append(batsman_stats, ignore_index=True)
```
