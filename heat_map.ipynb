```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import matplotlib
import io
import mplsoccer
```

```python
url = "https://raw.githubusercontent.com/Chaitanya98/Football-Analytics/main/Mad%20About%20Sports/Data/Real%20Madrid%20vs%20FC%20Barcelona%202015-16.csv"
dd = requests.get(url).content
df = pd.read_csv(io.StringIO(dd.decode('utf-8')))
```

```python
df.head(50)
```

```python
df.head(50)
```

```python
RMdf = df[(df['team']=="Real Madrid")&(df['outcomeType']=="Successful")&(df['playerName']=="Gareth Bale")]
```

```python
ccmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom cmap', ['green','red'])
pitch = mplsoccer.Pitch(pitch_type='opta', pitch_color ='green', line_color='white', line_zorder = 2)
fig,ax= pitch.draw(figsize=(14, 6))

pitch.kdeplot(RMdf['x'],RMdf['y'],ax=ax,cmap = ccmap, shade = True, n_levels =100, zorder = 2)
fig.suptitle("Gareth Bale Heat Map")
plt.show()
```
